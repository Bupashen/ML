```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('adult.data.csv')
```

```python
data
```

```python
# 1. How many men and women (sex feature) are represented in this dataset?
male_amount = len(data[data["sex"] == "Male"])
female_amount = len(data[data["sex"] == "Female"])
print(f"Male amount: {male_amount}")
print(f"Female amount: {female_amount}")
```

```python
#2. What is the average age (age feature) of women?
female = data[data["sex"] == "Female"]
means_of_feat = female.mean()
age_mean = means_of_feat["age"]
print(f"The average age of women: {age_mean}")
```

```python
#3. What is the percentage of German citizens (native-country feature)?
german = len(data[data["native-country"] == "German"])
overall = len(data)
print(f"The percentage of German citizens: {(german / overall)}%")
```

```python
#4-5. What are the mean and standard deviation of age 
#for those who earn more than 50K per year (salary feature) 
#and those who earn less than 50K per year?
data.groupby(["salary"])[["age"]].agg([np.mean, np.std])
```

```python
#6. Is it true that people who earn more than 50K have at least high school education? 
#(education â€“ Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters or Doctorate feature)
more_than_50k_edu = data[data["salary"] == ">50K"]["education"]
more_than_50k_edu_stats = more_than_50k_edu.value_counts()
print(more_than_50k_stats)
print("No, it's not true")
```

```python
#7. Display age statistics for each race (race feature) and each gender (sex feature). 
#Use groupby() and describe(). Find the maximum age of men of Amer-Indian-Eskimo race.
grouping_columns = ["race", "sex"]
columns_to_show = ["age"]
print(f"The maximum age of men of Amer-Indian-Eskimo race: {82.0}")
data.groupby(grouping_columns)[columns_to_show].describe()
```

```python
#8. Among whom is the proportion of those who earn a lot (>50K) greater: 
#married or single men (marital-status feature)? Consider as married those who 
#have a marital-status starting with Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), 
#the rest are considered bachelors.
marital_status_set = set(data["marital-status"])
married_status = set(["Married-civ-spouse", "Married-spouse-absent", "Married-AF-spouse"])
single_status = marital_status_set - married_status

married_men = data[data["marital-status"].isin(married_status)]
married_men_more_than_50k = married_men[married_men["salary"] == ">50K"]
married_men_more_than_50k_perc = len(married_men_more_than_50k) / len(married_men)

single_men = data[data["marital-status"].isin(single_status)]
single_men_more_than_50k = single_men[single_men["salary"] == ">50K"]
single_men_more_than_50k_perc = len(single_men_more_than_50k) / len(single_men)

print(f">50K among marriedmen: {married_men_more_than_50k_perc}")
print(f">50K among single men: {single_men_more_than_50k_perc}")
```

```python
#9. What is the maximum number of hours a person works per week (hours-per-week feature)? 
#How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?
max_hours = data["hours-per-week"].max()
print(f"The maximum number of hours: {max_hours}")
workholiks = data[data["hours-per-week"] == max_hours]
workholiks_amount = len(workholiks)
workholiks_more_than_50k = workholiks[workholiks["salary"] == ">50K"]
workholiks_more_than_50k_amount = len(workholiks_more_than_50k)
print(f"{workholiks_amount} people work such a number of hours.")
print(f"The percentage of those who earn a lot: {workholiks_more_than_50k_amount / workholiks_amount}")
```

```python
#10. Count the average time of work (hours-per-week) for those who earn a little and a lot (salary) 
#for each country (native-country). What will these be for Japan?
grouping_columns = ["native-country", "salary"]
columns_to_show = ["hours-per-week"]
print(data.groupby(grouping_columns)[columns_to_show].mean())
print(data[data["native-country"] == "Japan"].groupby(grouping_columns)[columns_to_show].mean())
```
